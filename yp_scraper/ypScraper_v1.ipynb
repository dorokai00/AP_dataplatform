```python
def landkreisfinder():
    #go to start page and nake into soup html obj
    url = "https://www.gelbeseiten.de/branchenbuch/staedte/schleswig-holstein/landkreise"
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')
    
    #find all landkreise
    lk = soup.find_all(class_="filterlist__item")
    landkreis_stack = [] #create empty stack to put landkreise into (not a stack tho)
    for e in lk:
        landkreis_stack.append(e.text.lower())

    return landkreis_stack
```

```python
def staedtefinder(landkreis_in_question):
    #pass relevant landkreis as argument and get its cities
    url = f"https://www.gelbeseiten.de/branchenbuch/staedte/schleswig-holstein/{landkreis_in_question}"
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')

    places = soup.find_all(class_="boxteaser__title")

    staedte_stack = []
    for e in places:
        staedte_stack.append(e.text.lower())

    return staedte_stack
```

```python
#works with https://www.gelbeseiten.de/branchenbuch/staedte/schleswig-holstein/{landkreis}/{stadt}/unternehmen
def unternehmensfinder(lk,stadt,unternehmensliste):
    url = f"https://www.gelbeseiten.de/branchenbuch/staedte/schleswig-holstein/{lk}/{stadt}/unternehmen"
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')

    all_links = soup.find_all('a',class_='link')

    for elm in all_links:
        #for each element in list of alle unternehmen per stadt in landkreis
        link_yp = elm.get("href") #get the yellowpage link for 1 buisness
        html = requests.get(link_yp).text #go there
        soup = BeautifulSoup(html, 'html.parser')

        #there are some very informative json structures there get those for now, since maybe more than one iterate again
        json_soup = soup.find_all(type="application/ld+json") #same as above
        for e in json_soup:
            e = e.text
            data = json.loads(e)

            #collect some information for now
            addr = data.get("address") #get adress
            keywords = data.get("keywords") #get keywords associated with buisness
            tel = data.get("telephone")
            openingHrs = data.get("openingHours")
            website = data.get("sameAs")
            description = data.get("description")
            lat, long= data.get("latitude"), data.get("longitude")

        unternehmensliste.append({"name":elm.get_text(), "link_yellowpages":link_yp,"address":addr,"keywords":keywords,"tel.":tel,
                                        "openingHours":openingHrs,"website":website,"description":description,"coordinates":(lat,long)})

    #return unternehmensliste
    
```
